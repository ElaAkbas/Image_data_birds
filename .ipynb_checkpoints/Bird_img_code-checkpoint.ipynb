```python
#import modules

import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os 
%matplotlib inline
from PIL import Image
from numpy import asarray
import numpy as np 
import csv
def change(string):
    result = ''
    for i in range(len(string)):
        if string[i] == "\\":
            result += '/'
        else:
            result += string[i]
    return result[2:] if result[0] == 'C' else result        
```

```python
bird_img_datalist = [['file', 'label', 'data_set']] #creating a header(s) for the final csv files

files_dir = change(r'C:\Users\elanu\Desktop\data\Bird images (kaggle)') #changing directory with function \ > /
os.chdir(files_dir) #working in this directory

with open(files_dir + '/birds.csv', 'r') as csvfile: #open provided birds csv file for all images from train,test and valid
    filedatareader = csv.reader(csvfile, delimiter=',', quotechar='|') 
    for row in filedatareader:
        each_row = []   #per loop adding a row of: numpy, label and type of data_set(train,test or valid)
        fname = row[1]  #row[1] is under filepaths, create fname to create path under this line
        image_path = files_dir + '/' + fname #path of every image
        if image_path == '/Users/elanu/Desktop/data/Bird images (kaggle)/filepaths': #skip header line
            continue
        else:
            #img = mpimg.imread(image_path)
            image = Image.open(image_path) #open image
            image_array = np.array(image)  #make array of image
            each_row.append(image_array)   #add arrays to first column
            each_row.append(row[2])        #add labels to second column
            each_row.append(row[4])        #add data_set to third column
        bird_img_datalist.append(each_row) #add whole row to the datalist

os.chdir(r'C:\Users\elanu\Desktop\bird_rep')
with open('bird_img_data.csv', 'w') as f: #create bird_img_data.csv file
    write = csv.writer(f) 
    write.writerows(bird_img_datalist)    
```
